```python
PLACE_ID = "QtXHCsgZwAZiFETM9"
```

```python
import os
import requests
import pandas as pd
from dotenv import load_dotenv

load_dotenv()
API_KEY = os.getenv("API_KEY")

url = (
    f"https://maps.googleapis.com/maps/api/place/details/json"
    f"?place_id={PLACE_ID}"
    f"&fields=name,reviews"
    f"&key={API_KEY}"
)

response = requests.get(url)
data = response.json()

# Check for API errors
if data.get("status") != "OK":
    print(f"API Error: {data.get('status')}")
    if "error_message" in data:
        print(f"Message: {data['error_message']}")
    print("Full response:")
    print(data)
    exit()

if "result" not in data:
    print("Missing 'result' in response. Full response:")
    print(data)
    exit()

reviews = []
for review in data["result"].get("reviews", []):
    reviews.append({
        "author": review.get("author_name"),
        "rating": review.get("rating"),
        "time": review.get("relative_time_description"),
        "text": review.get("text")
    })

df = pd.DataFrame(reviews)
df.to_csv("reviews.csv", index=False)
print(f"Saved {len(reviews)} reviews to reviews.csv")
```

**Scraping from Google Maps via third-party API**

```python
from apify_client import ApifyClient
import pandas as pd
import os
from dotenv import load_dotenv

load_dotenv()

API_KEY = os.getenv("APIFY_TOKEN")

client = ApifyClient(API_KEY)

run_input = {
    "startUrls": [
        {
            "url": "https://www.google.com/maps/place/Sushi+San/@2.2404817,102.2323121,17z/data=!3m1!4b1!4m6!3m5!1s0x31d1f136a83cef7b:0xd3cf99aae2945026!8m2!3d2.2404817!4d102.2323121!16s%2Fg%2F11t2lx2l2m?entry=ttu&g_ep=EgoyMDI2MDYxMC4wIKXMDSoASAFQAw%3D%3D"
        }
    ],
    "maxReviews": 200,
    "language": "en"
}

run = client.actor("compass/google-maps-reviews-scraper").call(run_input=run_input)

dataset_id = run.default_dataset_id

items = []

for item in client.dataset(dataset_id).iterate_items():
    items.append({
        "name": item.get("name"),
        "author": item.get("reviewerName"),
        "rating": item.get("stars"),
        "text": item.get("text"),
        "date": item.get("publishedAtDate")
    })

df = pd.DataFrame(items)
df.to_csv("google_maps_reviews.csv", index=False, encoding="utf-8-sig")

print("Saved google_maps_reviews.csv")
```

**Scraping from Google Maps via third-party API (Adjust Format)**

```python
from apify_client import ApifyClient
import pandas as pd
import os
import csv
from dotenv import load_dotenv

load_dotenv()

API_KEY = os.getenv("APIFY_TOKEN")

client = ApifyClient(API_KEY)

run_input = {
    "startUrls": [
        {
            "url": "https://www.google.com/maps/place/Sushi+Mentai+@+Malim+Jaya/data=!4m2!3m1!1s0x0:0xcc8e5cc178ad85d0?sa=X&ved=1t:2428&hl=en-MY&ictx=111"
        }
    ],
    "maxReviews": 200,
    "language": "en"
}

run = client.actor("compass/google-maps-reviews-scraper").call(run_input=run_input)

dataset_id = run.default_dataset_id

items = []

for idx, item in enumerate(
        client.dataset(dataset_id).iterate_items(),
        start=1):

    items.append({
        "number": idx,
        "name": item.get("name", ""),
        "rating": item.get("stars", ""),
        "text": item.get("text", ""),
        "date": item.get("publishedAtDate", "")
    })

df = pd.DataFrame(
    items,
    columns=[
        "number",
        "name",
        "rating",
        "text",
        "date"
    ]
)

df.to_csv(
    "google_maps_reviews.csv",
    index=False,
    encoding="utf-8-sig",
    quoting=csv.QUOTE_ALL
)

print(f"Saved {len(df)} reviews to google_maps_reviews.csv")
```
